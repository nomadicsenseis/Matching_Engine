#Full modular code

```python
# Versión de búsqueda de atributos por BM25 y selección del más relevante por BERT. Falta matizar la elección del BRICK luego

# Required installations and imports for running the code in a Jupyter Lab
# Note: The code is using some libraries like Sentence Transformers, rank_bm25, and spacy.
# Please make sure you have installed these libraries before running the code.

'''pip install -U pip setuptools wheel
pip install -U spacy
!python -m spacy download es_core_news_sm
pip install -U sentence-transformers
pip install rank_bm25
pip install apache_beam'''

import io
import pandas as pd
import xml.etree.ElementTree as et
import os
from pathlib import Path
import glob
import pickle
import torch
import numpy as np
from sentence_transformers import SentenceTransformer
from sentence_transformers import util
from rank_bm25 import BM25L
import apache_beam as beam
import spacy
nlp = spacy.load("es_core_news_sm")

# Attributes
# The GPC .xml database is imported with this notebook.
# If missing, please refer to the GPC official website and proceed with the free download.
attValues_df = pd.read_xml('GPC as of November 2021 v20211209 ES.xml', xpath="//brick/attType/attValue")
model = 'sentence-transformers/paraphrase-albert-small-v2'
GPC_xml = et.parse('GPC as of November 2021 v20211209 ES.xml')
attValue_embedding = pickle.load(open("serialattValueembedding.p", "rb"))
rep_dict = {'Spacy chunk': {'Query--attValue': 'Brick'}}

# Uncomment the following lines to save or load the rep_dict dictionary from a pickle file.
# pickle.dump(rep_dict, open("rep_dict.p", "wb"))
# rep_dict = pickle.load(open("rep_dict.p", "rb"))


def fromVal_getBricks(GPC_xml, attValue):
    """Return the grandparent of attValue (bricks).

    Args:
        GPC_xml (xml.etree.ElementTree): Parsed XML of the GPC database.
        attValue (str): Attribute value to search for.

    Returns:
        list: A list of grandparent elements (bricks) that contain the given attValue.
    """
    bricks = []
    for brick in GPC_xml.findall(f'.//attValue[@text="{attValue}"]/../..[@text]'):
        tag = 'text'
        attrib = brick.attrib
        text = attrib[tag]
        bricks.append(text)
    return bricks


def embedder(transformer_model, sentences):
    """Encode the given sentences using a Sentence Transformer model.

    Args:
        transformer_model (str): The Sentence Transformer model to use.
        sentences (list): List of sentences to be encoded.

    Returns:
        np.array: An array of sentence embeddings.
    """
    model = SentenceTransformer(transformer_model, device='cuda')
    embedding = model.encode(sentences)
    return embedding


def compare(corpus, query_embedding, corpus_embedding, metric, top_k, threshold):
    """Compare the query with the corpus using a given metric and return the top_k results.

    Args:
        corpus (list): List of sentences to compare the query against.
        query_embedding (np.array): Embedding representation of the query sentence.
        corpus_embedding (np.array): Embedding representations of the corpus sentences.
        metric (str): The metric to use for comparison ('cos_sim' for cosine similarity).
        top_k (int): Number of top results to return.
        threshold (float): The minimum similarity threshold to consider for a result.

    Returns:
        dict: A dictionary containing the top_k results with their corresponding similarity scores.
    """
    # Make sure a scalar product is used for comparison between the vectorized sentences.
    if metric == 'cos_sim':
        scores = util.cos_sim(query_embedding, corpus_embedding)[0]

    # Take the top_k results
    top_k = min(top_k, len(corpus_embedding))
    top_results = torch.topk(scores, k=top_k)

    # Make a dictionary with every sentence that is compared to the query and its score. Return that dictionary.
    top_res = {}
    for score, tensor in zip(top_results[0], top_results[1]):
        idx = tensor.item()
        if score.item() > threshold:
            top_res[corpus[idx]] = score.item()
    return top_res


def bm25(query, corpus, top_k):
    """Search for the given query in the corpus using BM25 and return the top_k results.

    Args:
        query (str): The query to search for in the corpus.
        corpus (list): List of sentences to search in.
        top_k (int): Number of top results to return.

    Returns:
        list: A list of top_k results found in the corpus using BM25.
    """
    query = query.upper()
    tokenized_corpus = [doc.split(' ') for doc in corpus]
    bm25 = BM25L(tokenized_corpus)

    tokenized_query = query.split(" ")
    doc_scores = bm25.get_scores(tokenized_query)
    results_query = bm25.get_top_n(tokenized_query, corpus, n=top_k)
    return results_query


def spacy(query):
    """Extract the product name from the given query using spaCy's noun_chunks.

    Args:
        query (str): The query from which to extract the product name.

    Returns:
        str: The product name extracted from the query.
    """
    doc = nlp(query)
    product = ''

    count = 0
    for chunk in doc.noun_chunks:
        count = count + 1
        if count < 3:
            if product == '':
                product = product + chunk.root.text
                if chunk.root.text != chunk.root.head.text:
                    product = product + ' ' + chunk.root.head.text
            else:
                if chunk.root.head.text in product:
                    product = product + ' ' + chunk.root.text
                else:
                    product = product + ' ' + chunk.root.text
                    if chunk.root.text != chunk.root.head.text:
                        product = product + ' ' + chunk.root.head.text
        else:
            break

    if product == '':
        product = doc[0].text
    return product

# Classifier class definition
class Classifier(beam.DoFn):

    def __init__(self, top_k, delimiter=','):
        """Initialize the Classifier class.

        Args:
            top_k (int): Number of top results to consider.
            delimiter (str): Delimiter to split the input text (queries).
        """
        self.delimiter = delimiter
        self.top_k = top_k
        self.model = model
        self.attValues_df = attValues_df
        self.attValue_embedding = attValue_embedding
        self.rep_dict = rep_dict

    def process(self, text):
        """Process the input text (queries) and return the top-k results for each query.

        Args:
            text (str): Input text containing queries separated by the delimiter.

        Yields:
            str: Comma-separated string of each query and its corresponding top-k results.
        """
        # Iterate through the queries in the input text
        for query in text.split(self.delimiter):

            # Clean query with spaCy
            spa_query = spacy(query)

            # Corpus of unique attributes
            aVU = self.attValues_df['text'].unique().tolist()

            # Check if spa_query has already been processed and exists in rep_dict
            top_brick = {}
            if spa_query in self.rep_dict.keys():
                top_brick[query] = self.rep_dict[spa_query]
            else:
                # Add spa_query to rep_dict if not already present
                self.rep_dict[spa_query] = {}

                # Query vectorization using the Sentence Transformer model
                spa_query_embedding = embedder(self.model, spa_query)

                # Compare query with attValues using BERT-based model (cosine similarity)
                top_res_BERT = compare(aVU, spa_query_embedding, self.attValue_embedding, 'cos_sim', self.top_k, 0.5)

                # Compare query with attributes using tf-idf based model (BM25)
                top_res_bm25 = bm25(spa_query, aVU, self.top_k)

                # Take only the attributes in the intersection if there are any, otherwise take the ones from BERT
                top_res = {}
                for attValue, score in top_res_BERT.items():
                    if attValue in top_res_bm25:
                        top_res[attValue] = score
                if not top_res:
                    # If no intersection, check synonyms or word embeddings and use top_res_BERT
                    top_res = top_res_BERT

                # Get bricks from attValues, vectorize them, and compare them with the query
                if top_res:

                    # Second architecture: Look for the most common brick
                    top_rep_bricks = {}
                    for attValue, score in top_res.items():
                        top_bricks = fromVal_getBricks(GPC_xml, attValue)
                        for brick in top_bricks:
                            if brick in top_rep_bricks:
                                top_rep_bricks[brick] = top_rep_bricks[brick] + score
                            else:
                                top_rep_bricks[brick] = score

                    top_rep_bricks_embedding = embedder(self.model, list(top_rep_bricks.keys()))
                    top_bricks_BERT = compare(list(top_rep_bricks.keys()), spa_query_embedding, top_rep_bricks_embedding, 'cos_sim', self.top_k, 0)
                    top_bricks = {k: top_rep_bricks.get(k, 0) + top_bricks_BERT.get(k, 0) for k in set(top_rep_bricks) | set(top_bricks_BERT)}

                    # Aquí falta saber cuando dos bricks tienen la misma puntuación
                    maxi = heapq.nlargest(self.top_k, top_bricks, key=top_bricks.get)
                    top_brick = {query: maxi}
                    self.rep_dict[spa_query] = maxi

                else:
                    top_brick[query] = 'No result found'
                    self.rep_dict[spa_query] = 'No result found'

            # Create a comma-separated string of the query and its corresponding top-k results
            top_brick_csv = f'{list(top_brick.keys())[0]},{list(top_brick.values())[0]}'

            yield top_brick_csv

# Pipeline definition using Apache Beam
with beam.Pipeline() as pipeline:
    products = (
        pipeline
        # Read the data from a csv file and extract the third column as text.
        | "Create df from csv" >> beam.io.ReadFromText('productos_para_métrica.csv', skip_header_lines=True)
        | 'map' >> beam.Map(lambda x: x.split(';')[2])
        # Use the Classifier class to process each query and get top-k results.
        | 'Classifier' >> beam

```
